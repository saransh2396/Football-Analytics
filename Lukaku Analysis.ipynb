```python
import pandas as pd
pd.set_option("display.max_columns", None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Lu_current = pd.read_csv('Possession Lukaku (2021-2022).csv')
Lu_lastyear = pd.read_csv('Lukaku F17 (2020-2021).csv')
```

```python
Lu_current.drop(['Unnamed: 0'],axis=1,inplace=True)
Lu_lastyear.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Lu_current
```

```python
Lu_lastyear
```

```python
from soccerplots.radar_chart import Radar
```

```python
new = pd.concat([Lu_lastyear,Lu_current]).reset_index()
new
```

```python
#dropping columns by col number
new.drop(new.columns[[0,1,2,3,4,5,7,8,9,12,16,17,18,19,21,22,25,26,27,28,-6,-3,-2,-1]],axis=1,inplace=True)
```

```python
new
```

```python
params = list(new.columns)
#params = required_columns[1:]
params = params[1:]
params
```

```python
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(new[params][x])
    a = a - (a*.25)
    
    b = max(new[params][x])
    b = b + (b*.25)
    
    ranges.append((a,b))
    
for x in range(len(new['Squad'])):
    if new['Squad'][x] =='Inter':
        a_values = new.iloc[x].values.tolist()
    if new['Squad'][x] == 'Chelsea':
        b_values = new.iloc[x].values.tolist()
        
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values,b_values]
```

```python
a_values
```

```python
title = dict(
    title_name='Inter',
    title_color = 'red',
    subtitle_name = 'Lukaku',
    subtitle_color = 'red',
    title_name_2='Chelsea',
    title_color_2 = 'blue',
    subtitle_name_2 = 'Lukaku',
    subtitle_color_2 = 'blue',
    title_fontsize = 18,
    subtitle_fontsize=15
)
```

```python
endnote = "Visualization made by: Saransh"
```

```python
radar = Radar(background_color='#A9DF8C',label_color='black',label_fontsize=14,range_fontsize=8,)
fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                         radar_color=['red','blue'],
                         alphas=[.85,.7],title=title,endnote=endnote,dpi=400,
                         compare=True)
plt.savefig('Lukaku_possession_comparision.png')
```
